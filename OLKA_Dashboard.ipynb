## 3️⃣ ngrok Token'ı Ayarla

```python
from pyngrok import ngrok, conf

# ngrok token'ı buraya yapıştır
NGROK_TOKEN = "your_ngrok_token_here"  # https://dashboard.ngrok.com/auth/your-authtoken adresinden al

if NGROK_TOKEN == "your_ngrok_token_here":
    print("⚠️  ngrok token'ı ayarla!")
    print("https://dashboard.ngrok.com adresine git ve token'ını kopyala")
else:
    ngrok.set_auth_token(NGROK_TOKEN)
    print("✅ ngrok hazır")
```

## 4️⃣ Dashboard'u Başlat

```python
import subprocess
import time
from pyngrok import ngrok
import json

# FastAPI sunucusunu başlat (arka planda)
process = subprocess.Popen(
    ["python", "-m", "uvicorn", "backend:app", "--host", "0.0.0.0", "--port", "8000"],
    stdout=subprocess.PIPE,
    stderr=subprocess.PIPE
)

# Sunucunun başlaması için bekle
time.sleep(3)

# ngrok tüneli oluştur
public_url = ngrok.connect(8000)

print("\n" + "="*60)
print("✅ OLKA Sprint Dashboard Çalışıyor!")
print("="*60)
print(f"\n🌐 Dashboard URL: {public_url}")
print("\n💡 İPUCU: URL'yi tarayıcınızda açın ve canlı dashboard'u kullanın!")
print("\n🔄 Dashboard otomatik olarak her 30 saniyede güncellenir.")
print("\n⏹️  Durdurmak için: aşağıda 'Çalıştırmayı İndir' butonuna tıklayın")
print("="*60)

# Başarı durumunu kontrol et
try:
    import requests
    time.sleep(1)
    response = requests.get(f"{public_url}/api/sprint", timeout=5)
    if response.status_code == 200:
        data = response.json()
        print(f"\n📊 Sprint: {data['sprint']['name']}")
        print(f"📈 Toplam İşler: {data['stats']['total']}")
        print(f"✅ Tamamlanan: {data['stats']['phases'].get('TAMAM', 0)}")
        print(f"🔄 Devam Eden: {data['stats']['phases'].get('DEVAM', 0)}")
        print(f"⚠️  Risk: {data['stats']['phases'].get('RISK', 0)}")
except Exception as e:
    print(f"\n⚠️  API kontrol başarısız: {e}")
```
